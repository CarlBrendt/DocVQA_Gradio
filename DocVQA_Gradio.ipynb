### Create Translation Function


```python
import requests
url = "https://deep-translate1.p.rapidapi.com/language/translate/v2"

def get_trans(text, to, target):
    payload = {
      "q": text,
      "source": to,
      "target": target
    }
    headers = {
      "content-type": "application/json",
      "X-RapidAPI-Key": "86c49ca3e3mshcc231781cb4055fp1adc90jsn6ebc802edeaf",
      "X-RapidAPI-Host": "deep-translate1.p.rapidapi.com"
    }

    response = requests.post(url, json=payload, headers=headers)

    data = response.json()
    return data['data']['translations']['translatedText']
```

```python
def translate_en2ru(text):
    return get_trans(text, "en", "ru")

def translate_ru2en(text):
    return get_trans(text, "ru", "en")
```

```python
from transformers import AutoProcessor, BlipForQuestionAnswering
import torch

blip_processor_base = AutoProcessor.from_pretrained("Salesforce/blip-vqa-base")
blip_model_base = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
blip_model_base.to(device)
```

```python
def generate_answer_blip(processor, model, image, question):

    inputs = processor(images=image, text=question, return_tensors="pt")

    generated_ids = model.generate(**inputs, max_length=50)
    generated_answer = processor.batch_decode(generated_ids, skip_special_tokens=True)

    return generated_answer


def generate_answers(image, question):

    # Перевести вопрос с русского на английский (если необходимо)
    question = translate_ru2en(question)

    # Здесь выполняйте генерацию ответа
    answer_blip_base = generate_answer_blip(blip_processor_base, blip_model_base, image, question)

    # Перевести ответ с английского на русский (если необходимо)
    answer_blip_base = translate_en2ru(answer_blip_base)

    return answer_blip_base
```

```python
import gradio as gr
import torch.hub

torch.hub.download_url_to_file('http://images.cocodataset.org/val2017/000000039769.jpg', 'cats.jpg')
torch.hub.download_url_to_file('https://huggingface.co/datasets/nielsr/textcaps-sample/resolve/main/stop_sign.png', 'stop_sign.png')
torch.hub.download_url_to_file('https://cdn.openai.com/dall-e-2/demos/text2im/astronaut/horse/photo/0.jpg', 'astronaut.jpg')

examples = [["cats.jpg", "How many cats are there?"], ["stop_sign.png", "What's behind the stop sign?"], ["astronaut.jpg", "What's the astronaut riding on?"]]
outputs = [gr.Textbox(label="Answer generated by BLIP-base")]

title = "Interactive demo: visual question answering (VQA) models"
description = "Gradio Demo BLIP , state-of-the-art vision+language models. To use it, simply upload your image and click 'submit', or click one of the examples to load them. Read more at the links below."
article = "<p style='text-align: center'><a href='https://huggingface.co/docs/transformers/main/model_doc/blip' target='_blank'>BLIP docs</a></p>"

from queue import Queue

# Initialize a queue
request_queue = Queue()

# Function to handle requests and put them in the queue
def handle_request(image, question):
    request_queue.put((image, question))

# Function to process requests from the queue
def process_requests():
    while True:
        image, question = request_queue.get()
        # Process the data here, image, and question are available for processing

# Start a separate thread to process requests
import threading
processing_thread = threading.Thread(target=process_requests)
processing_thread.start()

# Create a Gradio interface
interface = gr.Interface(
    fn=generate_answers,
    inputs=[gr.Image(type="pil"), gr.Textbox(label="Question")],
    outputs=outputs,
    examples=examples,
    title=title,
    description=description,
    article=article,
)
interface.launch(debug=True)
```

```python
import torch.hub
import speech_recognition as sr
import tempfile
from pydub import AudioSegment
from gtts import gTTS

def speech_to_text(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)
    text = recognizer.recognize_google(audio)
    return text

def text_to_speech(text, audio_file):
    tts = gTTS(text=text, lang='ru')
    tts.save(audio_file)

def process_audio(audio):
    audio_segment = AudioSegment.from_file(audio)
    with tempfile.NamedTemporaryFile(suffix=".wav") as temp:
        audio_segment.export(temp.name, format="wav")
        text = speech_to_text(temp.name)
    return text

def handle_request(image, question, audio):
    text = process_audio(audio)
    answers = generate_answers(image, text)
    answer_text = answers[0] if len(answers) > 0 else "No answer found"
    audio_file = audio.split(".")[0] + ".mp3"
    text_to_speech(answer_text, audio_file)
    return answer_text, audio_file

outputs = [
    gr.Textbox(label="Answer generated by BLIP"),
    gr.Audio(label="Answer audio")
]

torch.hub.download_url_to_file('http://images.cocodataset.org/val2017/000000039769.jpg', 'cats.jpg')
torch.hub.download_url_to_file('https://huggingface.co/datasets/nielsr/textcaps-sample/resolve/main/stop_sign.png', 'stop_sign.png')
torch.hub.download_url_to_file('https://cdn.openai.com/dall-e-2/demos/text2im/astronaut/horse/photo/0.jpg', 'astronaut.jpg')

examples = [["cats.jpg", "How many cats are there?"], ["stop_sign.png", "What's behind the stop sign?"], ["astronaut.jpg", "What's the astronaut riding on?"]]
#outputs = [gr.Textbox(label="Answer generated by BLIP-base")]

title = "Interactive demo: visual question answering (VQA) models"
description = "Gradio Demo BLIP , state-of-the-art vision+language models. To use it, simply upload your image and click 'submit', or click one of the examples to load them. Read more at the links below."
article = "<p style='text-align: center'><a href='https://huggingface.co/docs/transformers/main/model_doc/blip' target='_blank'>BLIP docs</a></p>"

from queue import Queue
# Initialize a queue
request_queue = Queue()

# Function to process requests from the queue
def process_requests():
    while True:
        image, question = request_queue.get()
        # Process the data here, image, and question are available for processing

# Start a separate thread to process requests
import threading
processing_thread = threading.Thread(target=process_requests)
processing_thread.start()

inputs = [
    gr.Image(type="pil"),
    gr.Textbox("Вопрос"),
    gr.Audio(type="filepath", label="Голосовой вопрос")
]

interface = gr.Interface(
    fn=handle_request,
    inputs=inputs,
    outputs=outputs,
    examples=examples,
    title=title,
    description=description,
    article=article,
)
interface.launch(debug=True)
```
